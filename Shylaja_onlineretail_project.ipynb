```python
# K-means clustering


from sklearn.preprocessing import StandardScaler

features = ['NumPurchases', 'TotalQuantity', 'TotalSpend', 'AvgUnitPrice']
X = customer_summary[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Use elbow method to choose K



from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertias = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Plot the Elbow
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertias, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
customer_summary['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
customer_summary['Cluster']
```

```python
customer_summary['Cluster'].value_counts()
```

```python
# Visualize Clusters Using PCA
# Reduce the dimensions from 4 to 2

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Add PCA coordinates to the dataframe
customer_summary['PCA1'] = X_pca[:, 0]
customer_summary['PCA2'] = X_pca[:, 1]

import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=customer_summary,
    x='PCA1', y='PCA2',
    hue='Cluster',
    palette='Set2',
    s=60,
    edgecolor='black'
)
plt.title('Customer Segments (2D PCA View)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
pca.components_
```

```python
pca_df = pd.DataFrame(pca.components_, columns=features, index=['PCA1', 'PCA2'])
print(pca_df)
```

```python
cluster_labels = {
    0: 'Average Shoppers',
    1: 'High-Value Customers',
    2: 'Regular Customers',
    3: 'Anomalous / Refund Cases'
}
```

```python
#mapping the customer_labels to clusters

customer_summary['Segment'] = customer_summary['Cluster'].map(cluster_labels)
```
